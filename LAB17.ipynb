```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df= pd.read_csv('loan_defaults.csv')
```

```python
df.head()
```

```python
df['EMPLOYMENT_TYPE'].value_counts().plot.bar()
plt.title('Employment Type')
```

```python
df['AVERAGE_ACCT_AGE'] = pd.DatetimeIndex(df['DISBURSAL_DATE']).year - pd.DatetimeIndex(df['DATE_OF_BIRTH']).year
df['AVERAGE_ACCT_AGE'].plot.hist()
plt.title('Age Histogram')
```

```python
sns.boxplot(x='NO_OF_INQUIRIES', y='AVERAGE_ACCT_AGE',data=df)
plt.title('Age BoxPlot')
```
